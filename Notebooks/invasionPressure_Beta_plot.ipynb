```python
def genFlName(mdl_name, kappa_val, N_val, B_val):
    kappa = kappa_val
    ku = "%.3E" % kappa
    ku = ku.replace(".","p")
    ku = ku.replace("-","m")
    ku = ku.replace("+","")

    N = N_val
    Nu = "%.3E" % N
    Nu = Nu.replace(".","p")
    Nu = Nu.replace("-","m")
    Nu = Nu.replace("+","")
    
    B = B_val
    Bu = "%.3E" % B
    Bu = Bu.replace(".","p")
    Bu = Bu.replace("-","m")
    Bu = Bu.replace("+","")

    flname = mdl_name + 'k_keq_' + ku + '_Neq_' +  Nu + '_Beq_' + Bu
    return flname
```
